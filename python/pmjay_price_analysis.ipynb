# PM-JAY vs Private Cardiac Package Price Analysis


```python
import pandas as pd
df = pd.read_excel('PM-JAY and Private hospitals price packages comparision.xlsx')
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
df.head()
```

```python
df.info()
df['group'].value_counts()
```

```python
price_summary = (
    df.groupby(['procedure_name', 'group'])['average_price']
      .mean()
      .reset_index()
)
price_summary
```

```python
pivot_table = price_summary.pivot(
    index='procedure_name',
    columns='group',
    values='average_price'
).reset_index()
pivot_table
```

```python
pivot_table['price_gap'] = pivot_table['PRIVATE'] - pivot_table['PM- JAY (INR)']
pivot_table['price_ratio'] = (pivot_table['PRIVATE'] / pivot_table['PM- JAY (INR)']).round(2)
pivot_table
```

```python
pivot_table.to_csv('pmjay_price_comparison_cleaned.csv', index=False)
```
